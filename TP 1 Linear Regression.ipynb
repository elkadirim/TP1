Une différence clé entre les deux bibliothèques est la façon dont elles gèrent les constantes. Scikit-learn permet à l'utilisateur de spécifier s'il faut ou non ajouter une constante via un paramètre, tandis que la classe OLS de statsmodels a une fonction qui ajoute une constante à un tableau donné.
Finalement, l'algorithme de sklearn est plus rapide pour des datasets plus larges
